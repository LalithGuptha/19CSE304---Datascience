```python
#14
m = len(df[(df.sex == " Male") & (df.income == " >50K")])
f = len(df[(df.sex == " Female") & (df.income == " >50K")])
print("Percentage of males is: ",end=" ")
perc=(m/(m+f))*100
print(perc)
```

```python
#15 skew and kurt0
temp=pd.DataFrame(df["age"])
print("Skew: ",temp.skew())
print("Kurtosis: ",temp.kurt())
```

```python
#16 propotion
male_high = len(df[(df.sex == " Male") & (df.income == " >50K")])
male = len(df[df.sex == " Male"])
print("Proportion of males with high income : ",end= " ")
print(male_high/male)
```
