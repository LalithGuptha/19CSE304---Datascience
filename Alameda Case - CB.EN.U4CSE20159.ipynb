```python
import numpy.random as npr
import numpy as np
npr.multinomial(100,[0.15,0.18,0.12,0.54,0.01])
```

```python
simulations = 1000
for i in np.arange(simulations):
    count_pop = npr.multinomial(100, [0.15,0.18,0.12,0.54,0.01], simulations)
count_pop[1:5]
```

```python
import pandas as pd
panel_df = pd.DataFrame(count_pop)
panel_df.head(3)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (7,7))
panel_df[0].hist()
plt.title('Statistic');
plt.xlabel('Count in a Random Sample % for Asian',fontsize=15)
plt.ylabel('Unit',fontsize=15);
plt.scatter(8, 0, color='red', s=30)
```
