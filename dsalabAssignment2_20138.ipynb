<h3> Answer 3 </h3>

```python
df=pd.read_csv("Question3.csv")
```

```python
df
```

```python
df["Age"].unique()
```

```python
df["Year"].unique()
```

```python
df=df[(((df["Age"]=="18 to 64") | (df["Age"]=="65 to 80+")) & (df['Year']=="1/1/14 0:00"))]
```

```python
df
```

<p> <b> (1) </b> </p>

```python
req=df.groupby("Educational Attainment").sum("Population Count")
req
```

<p> <b> (2) </b> </p>

```python
req["Percentage"]=(req["Population Count"]/req["Population Count"].sum())*100
```

```python
req
```

<p> <b> (3) </b> </p>

```python
new_df=df.groupby(['Educational Attainment','Personal Income'],as_index=False).sum("Population Count")
```

```python
new_df
```

```python
df_pivot=new_df.pivot(index='Personal Income',columns='Educational Attainment',values="Population Count")
df_pivot
```

<p> <b> (4) </b> </p>

```python
df_pivot=df_pivot.drop(columns=["High school or equivalent","College, less than 4-yr degree"])
df_pivot
```

```python
df_pivot.plot(kind="barh")
```
